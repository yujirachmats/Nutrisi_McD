```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('menu.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Serving Size'].unique()
```

```python
dfx = df.copy()
dfx.head()
```

## Cleaning Serving Size data

```python
hsl = []
for i in df['Serving Size']:
    # 236 ml == 7.98 oz
    if i == '1 carton (236 ml)':
        hsl.append(float(7.98))
    # 33g == 0.035 oz
    elif i == '1 cookie (33 g)':
        hsl.append(float(0.035))
    else:
        hsl.append(float(i.split(' ')[0]))
dfx['Serving Size'] = hsl
dfx.head()
```

## Mean,Median & Modus

- Untuk feature/column Serving Size, Calories, & Total Fat

```python
# mean
meanSs = dfx['Serving Size'].mean()
meanC = dfx['Calories'].mean()
meanTF = dfx['Total Fat'].mean()
# print(meanSs,meanC,meanTF)

# median
medianSs = dfx['Serving Size'].median()
medianC = dfx['Calories'].median()
medianTF = dfx['Total Fat'].median()
# print(medianSs, medianC, medianTF)

# mode
modeSs = dfx['Serving Size'].mode()
modeC = dfx['Calories'].mode()
modeTF = dfx['Total Fat'].mode()
print(modeSs[0],modeC[0],modeTF[0])
```

## Range, Q1, Q3 dan IQR

```python
# Range
rangedfx = dfx[['Serving Size','Calories','Total Fat']].max() - dfx[['Serving Size','Calories','Total Fat']].min()
rangeSs = rangedfx[0]
rangeC = rangedfx[1]
rangeTF = rangedfx[2]
print(rangeSs,rangeC,rangeTF)

# Q1, Q3, & IQR
q1Ss = dfx['Serving Size'].quantile(.25)
q3Ss = dfx['Serving Size'].quantile(.75)
q1C = dfx['Calories'].quantile(.25)
q3C = dfx['Calories'].quantile(.75)
q1TF = dfx['Total Fat'].quantile(.25)
q3TF = dfx['Total Fat'].quantile(.75)
print(q1Ss,q3Ss,q1C,q3C,q1TF,q3TF)

iqrSs = q3Ss - q1Ss
iqrC = q3C - q1C
iqrTF = q3TF - q1TF
print(iqrSs,iqrC,iqrTF)
```

## Std & Variance

```python
stdSs = dfx['Serving Size'].std()
stdC = dfx['Calories'].std()
stdTF = dfx['Total Fat'].std()
print(stdSs,stdC,stdTF)

varSs = dfx['Serving Size'].var()
varC = dfx['Calories'].var()
varTF = dfx['Total Fat'].var()
print(varSs,varC,varTF)
```

## Z-score tiap data point

```python
for col in dfx[['Serving Size','Calories','Total Fat']]:
    col_zscore = col + '_zscore'
    dfx[col_zscore] = (dfx[col] - dfx[col].mean())/dfx[col].std(ddof=0)
dfx[dfx.columns[24:]]
```

## IQR Method

```python
dfxIQR = dfx[['Serving Size','Calories','Total Fat']].quantile(.75) - dfx[['Serving Size','Calories','Total Fat']].quantile(.25)
highIQR = dfx[['Serving Size','Calories','Total Fat']].quantile(.75) + (1.5*dfxIQR)
lowIQR = dfx[['Serving Size','Calories','Total Fat']].quantile(.25) - (1.5*dfxIQR)
print(highIQR, lowIQR)
```

```python
iqrOutlierSs = dfx[(dfx['Serving Size'] <= lowIQR[0]) | (dfx['Serving Size'] >= highIQR[0])]
iqrOutlierC = dfx[(dfx['Calories'] <= lowIQR[1]) | (dfx['Calories'] >= highIQR[1])]
iqrOutlierTF = dfx[(dfx['Total Fat'] <= lowIQR[2]) | (dfx['Total Fat'] >= highIQR[2])]
iqrOutlierSs.index
```

## Z score Method

- high outlier = 2.5 & low outlier = -2.5 values

```python
zOutlierSs = dfx[(dfx['Serving Size'] <= -2.5) | (dfx['Serving Size'] >= 2.5)]
zOutlierC = dfx[(dfx['Calories'] <= -2.5) | (dfx['Calories'] >= 2.5)]
zOutlierTF = dfx[(dfx['Total Fat'] <= lowIQR[0]) | (dfx['Total Fat'] >= 2.5)]
zOutlierSs.index
```
